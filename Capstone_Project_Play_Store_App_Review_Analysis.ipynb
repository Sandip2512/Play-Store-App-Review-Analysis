## 3. ***Data Wrangling***

### Data Wrangling Code

````python
# Write your code to make your dataset analysis ready.

#Remove duplicates apps
merged_df.drop_duplicates(subset='App', inplace=True)

#Clean "Installs" column
merged_df.loc[:,'Installs'] = merged_df['Installs'].astype(str)
merged_df = merged_df[merged_df['Installs'].str.contains(r'^\d', na=False)]
merged_df.loc[:,'Installs'] = merged_df['Installs'].str.replace('[+,]', '', regex=True).astype(int)

#Clean Price
merged_df.loc[:,'Reviews'] = pd.to_numeric(merged_df['Reviews'], errors='coerce')

#Clean size
def size_mb(size):
  if pd.isna(size):
    return np.nan
  size = str(size)
  if size == 'Varies with device':
    return np.nan
  if 'M' in size:
    return float(size.replace('M', '').strip())
  if 'k' in size:
    return float(size.replace('k', '').strip()) / 1024
  return np.nan

merged_df.loc[:,'Size'] = merged_df['Size'].map(size_mb)

#Drop rows with missing essential fields
merged_df.dropna(subset=['Rating', 'Type','Reviews'], inplace=True)

#Clean user reviews
df_reviews = merged_df.dropna(subset=['Translated_Review', 'Sentiment'])
df_reviews = df_reviews.drop_duplicates()
df_reviews.reset_index(drop=True, inplace=True)
````

### What all manipulations have you done and insights you found?

Answere:

We cleaned the dataset by removing repeated rows, fixing incorrect entries in the Installs and Size columns, and deleting rows that had missing or empty values in important columns like reviews or sentiment. Now the data is neat and ready to use for analysis.
